# Tarea - Caso real de Big Data
La marca Rolls-Royce puede ser mayormente reconocida por los carros de lujo que fabrica, sin embargo, tambien fabrican motores enormes y potentes que impulsan aviones por los cielos y barcos por los mares. Estos motores y sistemas de propulsion vienen equipados con diversos sensores que registran el funcionamiento del componente y envian estos datos en tiempo real a los ingenieros quienes en base a estos datos decidiran que accion tomar. Con todo el volumen de datos que genera de estos motores y los instrumentos de analisis cada vez mas sofisticados se espera poder impulsar la eficiencia y el progreso. 

Rolls-Royce tiene en mente que en un futuro con el avance de esta tecnologia, poder automatizar barcos de carga que puedan viajar solos con la minima intervencion humana y el minimo margen de error, al poder tomar decisiones frente a circunstancias como las condiciones meteorologicas. 


### ¿Qué datos genera?
La empresa utiliza metodos de big data en tres areas clave de sus operaciones: diseño, fabricacion y servicio posventa. Genera grandes volumenes de datos en simulaciones de sus motores a reaccion. Saber como leer estos datos es igual de importante como saber manipular estos datos, reduce el tiempo de desarrollo y mejora la calidad y el rendimiento.

### ¿Quién los usa?
Estos datos son usados tanto por la empresa como por los clientes, por la empresa en el proceso de desarrollo al realizar simulaciones del componente en diversas circunstancias para determinar si el producto es bueno o malo. Por parte de los clientes al llevar un monitoreo del motor y poder darle seguimiento a la integridad y estado para saber cuando programar un mantenimiento o saber cuando algun detalle esta fallando.

###  ¿Qué problema resuelve?
Los problemas que resuelven es poder preveer fallos en los sistemas de propulsion antes de que pasen al saber en que momento realizar mantenimientos, poder estar seguros del desempeño que tendra el motor en diversos escenarios, poder denotar cuales son las areas que necesitan una mejora sustancial.

### ¿Por qué no bastaría Excel?
Paul Stein, director cientifico de la empresa comento en una entrevista para Forbes, "Producimos en nuestra fabrica de Singapur, medio terabyte de datos de fabricacion por cada aspa de ventilador y producimos 6000 aspas al año, lo que supone tres petabytes de datos sobre la fabricacion de un solo componente.", es una cantidad enorme que seria imposible de mantener con las limitaciones que posee excel.
