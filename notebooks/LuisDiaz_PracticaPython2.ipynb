# Ejercicio 16:   Dataframe de estudiantes

### Enunciado: Cree un DataFrame con al menos 10 estudiantes, incluya nombre, carrera, edad, promedio y ciudad, muestre las primeras 5 filas, muestre la información general del DataFrame. Calcule el promedio general de calificaciones, muestre estudiantes con promedio mayor o igual a 80 y ordene estudiantes por promedio de mayor a menor.

### Lógica: se crea primero un diccionario de datos que se transforma en un DataFrame usando la extension panda, para calcular el promedio entre los estudiantes, se especifica la columna de interes y luego se usa la funcion describe() que calcula, entre otras cosas, el promedio de la serie de numeros, luego al imprimir se selecciona la fila de interes a mostrar. Para filtrar el promedio de los estudiantes, se crea la variable donde almacenaremos ese DataFrame filtrado seleccionando la columna Promedio y definiendo la condicion mayor que 80. Y por ultimo para ordenar de mayor a menor, aplicamos la funcion sort_values() le pasamos como parametro la columna que tomara para ordenar asignamos el valor False a ascending para que muestra el dataframe de manera descendente.

```python
import pandas as pd

estudiantes = {
    "Nombre": ["Julian", "Gonzalo", "Sofia", "Valentina", "Matias", "Camila", "Federico", "Isabella", "Santiago", "Lucia"],
    "Carrera": ["Ingeniería", "Medicina", "Derecho", "Arquitectura", "Psicología", "Economía", "Biología", "Comunicación", "Física", "Química"],
    "Edad": [20, 22, 21, 23, 19, 24, 22, 21, 20, 23],
    "Promedio": [74, 90, 78, 82, 95, 80, 75, 88, 92, 79],
    "Ciudad": ["San Pedro Sula", "Tegucigalpa", "La Ceiba", "Choluteca", "Comayagua", "Santa Rosa de Copán", "Danlí", "Tela", "El Progreso", "Juticalpa"]
}

df = pd.DataFrame(estudiantes)
promedio = df["Promedio"].describe() # calcula el promedio de la columna
filtroPromedio = df[df["Promedio"]  >= 80] # filtra los estudiantes con promedio mayor a 80
ordenadoPromedio = df.sort_values(by="Promedio", ascending=False)

print(df)

print("\n###Primeras 5 filas: ###\n\n" + str(df.head())) # muestra las primeras filas del DataFrame

print("\n###Promedio de los 10 estudiantes: ###  " + str(promedio.iloc[1])) # muestra el promedio

print("\n###Filtrado por estudiantes con promedio >= 80: ###\n\n" + str(filtroPromedio))

print("\n###Estudiantes ordenados de mayor a menor promedio: ###\n\n" + str(ordenadoPromedio))
```

# Ejercicio 17:   Análisis de ventas con pandas

### Enunciado: Cree un DataFrame de ventas con al menos 15 registros, incluya fecha, vendedor, producto, categoría, cantidad y precio unitario. Calcule una nueva columna TotalVenta, calcule el total vendido por vendedor, identifique el producto más vendido, identifique la categoría con mayor ingreso y ordene las ventas de mayor a menor.

### Lógica: Se crea el diccionario con los 15 registros, para calcular la nueva columna simplemente se crea escribiendo el nombre del data frame junto con el nombre de la columna a crear, y se le asigna su valor que, en este caso, es la multiplicación de la columna precio y cantidad. Para calcular el total vendido por vendedor usamos la función groupby y ponemos la columna a agrupar, luego le decimos que queremos sumar la columna TotalVenta con la función sum(). Para identificar el producto mas vendido usamos la misma sintaxis que la anterior, con la función groupby agrupamos la columna Producto y sumamos la cantidad, y agregamos la función idxmax() para conocer el valor máximo de ese resultado. Lo mismo hacemos para identificar la categoria con mas ingresos y finalmente para ordenar las ventas utilizamos la función sort_values, le decimos que ordenamos por TotalVenta y de manera descendiente.

```python
import pandas as pd 

ventas = {
    "Producto": ["Laptop", "Teclado", "Smartphone", "Tablet", "Monitor", "Teclado", "Smartphone", "Impresora", "Auriculares", "Smartphone", "Disco Duro", "Teclado", "Impresora", "Laptop", "Auriculares"],
    "Precio": [1200, 400, 800, 300, 150, 400, 800, 100, 70, 800, 100, 400, 70, 1200, 70],
    "Cantidad": [10, 20, 15, 8, 50, 100, 5, 25, 12, 30, 10, 20, 15, 10, 25],
    "Categoría": ["Electrónica", "Periféricos", "Electrónica", "Electrónica", "Periféricos", "Periféricos", "Electrónica", "Electrónica", "Audio", "Electrónica", "Almacenamiento", "Periféricos", "Electrónica", "Electrónica", "Audio"],
    "Fecha": ["2024-01-15", "2024-01-16", "2024-01-17", "2024-01-18", "2024-01-19", "2024-01-20", "2024-01-21", "2024-01-22", "2024-01-23", "2024-01-24", "2024-01-25", "2024-01-26", "2024-01-27", "2024-01-28", "2024-01-29"],
    "Vendedor": ["Elena", "Ana", "Andrés", "Ana", "Ana", "Laura", "Diego", "Laura", "Diego", "Elena", "Lucía", "Laura", "Diego", "Lucía", "Elena"]
}

df_ventas = pd.DataFrame(ventas)

df_ventas["TotalVenta"] = df_ventas["Precio"] * df_ventas["Cantidad"] # calcula el total de cada venta

total_vendedor = df_ventas.groupby("Vendedor")["TotalVenta"].sum() # agrupa por vendedor y suma el total de ventas
producto_mas_vendido = df_ventas.groupby("Producto")["Cantidad"].sum().idxmax() # encuentra el producto más vendido
categoria_mas_ingresos = df_ventas.groupby("Categoría")["TotalVenta"].sum().idxmax() # encuentra la categoría con más ingresos
ventas_ordenadas = df_ventas.sort_values(by="TotalVenta", ascending=False) # ordena las ventas por total de venta de mayor a menor

print(df_ventas)

print("\n###Total de ventas por vendedor: ###\n\n" + str(total_vendedor))
print("\n###Producto más vendido: ###\n\n" + str(producto_mas_vendido))
print("\n###Categoría con más ingresos: ###\n\n" + str(categoria_mas_ingresos))
print("\n###Ventas ordenadas por total de venta: ###\n\n" + str(ventas_ordenadas))
```

# Ejercicio 18:   Limpieza de datos con pandas

### Enunciado: Cree un DataFrame con datos incompletos de clientes. Incluya nombre, edad, correo, ciudad y teléfono, identifique valores nulos, cuente los valores nulos por columna, reemplace ciudades vacías con "No especificada", elimine registros que no tengan correo y muestre el DataFrame limpio.


### Lógica: Para contar los valores nulos usamos la función isnull() en el dataframe que ira dato por dato verificando si es nulo, y usamos sum() para sumar los resultados de cada columna, para reemplazar los datos nulos usamos la función fillna() que reemplaza valores nulos por el parámetro que le pasemos y para descartar los registros que no tengan correo simplemente aplicamos una condición a la columna correo con la función notnull que verifica si el valor es o no es nulo.

```python
import pandas as pd


datos = {
    "Nombre": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Heidi"],
    "Edad": [25, 30, 35, 40, 45, 50, 55, 60],
    "Correo": ["alice@example.com", None, "charlie@example.com", "david@example.com", "eve@example.com", "frank@example.com", "grace@example.com", None],
    "Ciudad": ["New York", None, "Chicago", None, "Phoenix", None, "San Antonio", "San Diego"],
    "Telefono": ["1456-7890", "2345-6789", "3456-7890", "4567-8901", "5678-9012", "6789-0123", "7890-1234", "8901-2345"]
}   

df_datos = pd.DataFrame(datos)

print("### Dataframe original: ###\n\n" + str(df_datos))

valores_nulos = df_datos.isnull().sum() # cuenta los valores nulos en cada columna
df_datos["Ciudad"] = df_datos["Ciudad"].fillna("No especificada") # reemplaza los valores nulos con "No especificada"
df_datos = df_datos[df_datos["Correo"].notnull()] # filtra los datos para mostrar solo las filas donde el correo no es nulo

print("\n\n### Dataframe limpio: ###\n\n" + str(df_datos))

print("\n\n### Valores nulos por columna: ###\n\n" + str(valores_nulos))
```

# Ejercicio 19:   Análisis de préstamos

### Enunciado: Cree un DataFrame con solicitudes de préstamo, incluya cliente, monto solicitado, plazo, ingreso mensual y estado de solicitud, calcule el monto promedio solicitado, cuente solicitudes por estado. Muestre únicamente préstamos aprobados, muestre clientes con monto solicitado mayor a 50,000, agrupe por estado y calcule el promedio del monto solicitado.


### Lógica: Se calcula el monto promedio de todos los préstamos usando mean(), que obtiene la media de la columna “Monto”. Después se filtran los datos para obtener solo los préstamos cuyo Estado es “Aprobado”, usando una condición booleana dentro del DataFrame. También se cuenta cuántos préstamos fueron aprobados y rechazados usando shape[0], que devuelve la cantidad de filas. Más adelante se filtra los préstamos cuyo monto es mayor a 50000 aplicando otra condición sobre la columna “Monto”. Finalmente, se usa groupby("Estado") para agrupar los datos según si están aprobados o rechazados y calcula el promedio del monto en cada grupo con mean().

```python
import pandas as pd


solicitudes_prestamo = {
    "Cliente": ["Juan", "María", "Carlos", "Ana", "Luis", "Sofía", "Miguel", "Laura"],
    "Monto": [50000, 100000, 75000, 12000, 30000, 15000, 8000, 20000],
    "Plazo": [12, 24, 18, 36, 6, 48, 24, 12],
    "IngresoMensual": [2000, 4000, 3000, 5000, 1500, 6000, 3500, 2500],
    "Estado": ["Aprobado", "Rechazado", "Aprobado", "Rechazado", "Aprobado", "Aprobado", "Aprobado", "Rechazado"]
}

df_prestamos = pd.DataFrame(solicitudes_prestamo)

monto_promedio = df_prestamos["Monto"].mean() # calcula el monto promedio de los préstamos
aprobados = df_prestamos[df_prestamos["Estado"] == "Aprobado"] # filtra los préstamos aprobados
num_aprobados = df_prestamos[df_prestamos["Estado"] == "Aprobado"].shape[0] # cuenta los préstamos aprobados
num_rechazados = df_prestamos[df_prestamos["Estado"] == "Rechazado"].shape[0] # cuenta los préstamos rechazados
montoMayor = df_prestamos[df_prestamos["Monto"] > 50000] # filtra los préstamos con monto mayor a 50000
estadoMontoPromedio = df_prestamos.groupby("Estado")["Monto"].mean() # calcula el monto promedio por estado de aprobación

print("Monto promedio de los préstamos: " + str(monto_promedio))
print("Número de préstamos aprobados: " + str(num_aprobados))
print("Número de préstamos rechazados: " + str(num_rechazados))
print("Monto promedio por estado de aprobación:\n" + str(estadoMontoPromedio))

print("\n### Préstamos aprobados: ###\n\n" + str(aprobados))
print("\n### Préstamos con monto mayor a 50000: ###\n\n" + str(montoMayor))
```

# Ejercicio 20:   Mini análisis de datos tipo empresa

### Enunciado: Cree un DataFrame con información de empleados, incluya nombre, departamento, cargo, salario, años de experiencia y ciudad, muestre los primeros registros, muestre estadísticas generales de salarios, calcule el salario promedio por departamento. Encuentre el empleado con salario más alto, muestre empleados con más de 5 años de experiencia, ordene empleados por salario, cree una columna NivelExperiencia: Junior, Intermedio o Senior.


### Lógica: se obtienen los primeros registros del DataFrame con head(), que sirve para ver una muestra inicial de los datos. Después se calculan las estadísticas del salario usando describe(), que muestra valores como promedio, mínimo, máximo y percentiles. También se calcula el salario promedio por departamento usando groupby("Departamento") y mean(), agrupando a los empleados según su área de trabajo. Para encontrar al empleado con mejor salario se usa idxmax(), que devuelve el índice de la fila con el valor más alto en la columna “Salario”. El código también filtra empleados con más de 5 años de experiencia usando una condición booleana, y ordena a todos los empleados por salario de menor a mayor con sort_values(). Finalmente, crea una nueva columna llamada “NivelExperiencia” que asigna manualmente categorías como Junior, Senior e Intermedio, y muestra el DataFrame final con esa información añadida.

```python
import pandas as pd

empleados = {
    "Nombre": ["Ana", "Luis", "Sofía", "Carlos", "María", "Miguel", "Laura", "Juan"],
    "Cargo": ["Gerente", "Analista", "Coordinador", "Director", "Ejecutivo", "Analista", "Coordinador", "Director"],
    "Departamento": ["Ventas", "Marketing", "Recursos Humanos", "Finanzas", "Ventas", "Marketing", "Recursos Humanos", "Finanzas"],
    "Salario": [50000, 60000, 55000, 70000, 52000, 62000, 58000, 75000],
    "AñosExperiencia": [5, 7, 6, 10, 4, 3, 6, 12],
    "Ciudad": ["Madrid", "Barcelona", "Valencia", "Sevilla", "Madrid", "Barcelona", "Valencia", "Sevilla"]
}

df_empleados = pd.DataFrame(empleados)

primeros_registros = df_empleados.head()
estadisticas_empleados = df_empleados["Salario"].describe()
salario_promedio_departamento = df_empleados.groupby("Departamento")["Salario"].mean()
empleado_mejor_salario = df_empleados["Salario"].idxmax()
empleados_mayor_experiencia = df_empleados[df_empleados["AñosExperiencia"] > 5]
empleados_ordenados_salario = df_empleados.sort_values(by="Salario")

df_empleados["NivelExperiencia"] = ["Junior", "Senior", "Intermedio", "Senior", "Junior", "Senior", "Intermedio", "Senior"] 

print("### Primeros registros del DataFrame: ###\n" + str(primeros_registros))
print("\n### Estadísticas de salarios: ###\n" + str(estadisticas_empleados))
print("\n### Salario promedio por departamento: ###\n" + str(salario_promedio_departamento))
print("\n### Empleado con el mejor salario: ###\n" + str(df_empleados.loc[empleado_mejor_salario]))
print("\n### Empleados con más de 5 años de experiencia: ###\n" + str(empleados_mayor_experiencia))
print("\n### Empleados ordenados por salario: ###\n" + str(empleados_ordenados_salario))

print("\n### DataFrame con nivel de experiencia: ###\n" + str(df_empleados))
```
