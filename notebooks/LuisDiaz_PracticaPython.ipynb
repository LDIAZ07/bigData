# Ejercicio 1: Clasificación de edad

### Enunciado: dependiendo de que valor tenga la variable edad, se calificara si es niño, adolescente, adulto o adulto mayor

### Lógica: se usa una función clasifyAge para retornar el valor de la condicioón que sea cierta

```python
def clasifyAge(age):
    if age<13:
        return "Niño"
    if age>=13 and age<=17:
        return "Adolescente"
    if age>=18 and age<=59:
        return "Adulto"
    if age>=60:
        return "Adulto mayor"

age=10

print("La categoria es: "+clasifyAge(age))
```

# Ejercicio 3: Calculadora básica

### Enunciado: armar una calculadora que pueda sumar, restar, dividir y multiplicar

### Lógica: se definen dos numeros y segun lo que contenga la variable operation, se decide que operación aritmetica se realizara

```python
operation="Resta"
num1=15
num2=5

def calculateNumbers(num1,num2,operation):
    match operation:
        case "Suma":
            return num1+num2
        case "Resta":
            return num1-num2
        case "Multiplicar":
            return num1*num2
        case "Dividir":
            return num1/num2

print(calculateNumbers(num1,num2,operation))
```

# Ejercicio 6: Contador de vocales

# Ejercicio 7: Promedio de calificaciones

### Enunciado: Definir 5 calificaciones para calcular su promedio e indicar si el estudiante aprueba o reprueba

### Lógica: las calificaciones almacenadas en un arreglo se suman con la funcion sum para luego sacar el promedio con la funcion len, el resultado si es menor a 70 indicara que esta reprobado, de lo contrario estara aprobado

```python
calificaciones = [80, 60, 70, 60, 75]

suma = sum(calificaciones)
promedio = suma / len(calificaciones)

print(f"Suma total: {suma}")
print(f"Promedio: {promedio}")

if promedio >= 70:
    print("Estado: Aprobado")
else:
    print("Estado: Reprobado")
```

# Ejercicio 8: Lista de números positivos y negativos

### Enunciado: de una lista de numeros, contar cuantos son positivos. negativos y ceros

### Lógica: se recorre el arreglo de numeros y si cumple cierta condicion, ira contando si es positivo, negativo o cero

```python
numeros = [4, -2, 0, 7, -5, 3, 0, -1, 8, -3]

positivos = 0
negativos = 0
ceros = 0

for n in numeros:
    if n > 0:
        positivos += 1
    elif n < 0:
        negativos += 1
    else:
        ceros += 1

print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Ceros: {ceros}")
```

# Ejercicio 11:  Función para calcular descuento

### Enunciado: crear una función llamada calcular_descuento, debe recibir precio original y porcentaje de descuento, debe retornar el precio final

### Lógica: se crea un diccionario llamado estudiante, dentro de el se asignan el nombre de llave y su valor, luego para mostrar el dato deseado, se coloca el nombre de la llave para acceder a su valor

```python
def calcular_descuento(precio_original, porcentaje_descuento):
    descuento    = precio_original * (porcentaje_descuento / 100)
    precio_final = precio_original - descuento
    return precio_final

precio    = 500
descuento = 20

resultado = calcular_descuento(precio, descuento)
print("Precio original : L. " + str(precio))
print("Descuento       : " + str(descuento) + "%")
print("Precio final    : L. " + str(resultado))
```

# Ejercicio 12:  Validación de contraseña

### Enunciado: Valide una contraseña definida por el usuario o por el código, debe tener al menos 8 caracteres, contener al menos un número, contener al menos una letra mayúscula, contener al menos una letra minúscula, indique si la contraseña es válida o no.

### Lógica: el codigo valida una contraseña verificando varias condiciones en orden. Primero revisa si tiene al menos 8 caracteres, y si no cumple esa condición la rechaza de inmediato. Luego recorre la contraseña para comprobar si contiene al menos un número; si no encuentra ninguno, también la marca como invalida. Después hace otra revisión para ver si existe al menos una letra mayúscula, y si no la tiene, la rechaza. Finalmente verifica que haya al menos una letra minúscula. Si la contraseña cumple con todas estas reglas, entonces se considera válida y el programa lo indica con un mensaje de aceptación

```python
def validar_contrasena(contrasena):
    if len(contrasena) < 8:
        return "NO válida: mínimo 8 caracteres."

    tiene_numero = False
    for caracter in contrasena:
        if caracter.isdigit():
            tiene_numero = True
    if tiene_numero == False:
        return "NO válida: debe tener al menos un número."

    tiene_mayuscula = False
    for caracter in contrasena:
        if caracter.isupper():
            tiene_mayuscula = True
    if tiene_mayuscula == False:
        return "NO válida: debe tener mayúscula."

    tiene_minuscula = False
    for caracter in contrasena:
        if caracter.islower():
            tiene_minuscula = True
    if tiene_minuscula == False:
        return "NO válida: debe tener minúscula."

    return "Contraseña VÁLIDA. Cumple todos los requisitos."

print(validar_contrasena("abc123"))   
print(validar_contrasena("abcdefgh"))
print(validar_contrasena("ABCDEFG1")) 
print(validar_contrasena("MiClave9")) 
```

# Ejercicio 13:  Sistema simple de inventario

### Enunciado: Cree una lista de productos, cada producto debe tener código, nombre, precio y cantidad disponible. Muestre todos los productos, busque un producto por código y calcule el valor total del inventario

### Lógica: el código crea un inventario de productos usando pandas y lo organiza en una tabla. Luego muestra todos los productos y busca uno específico por su código. Después calcula el valor total de cada producto multiplicando precio por cantidad, y suma todo el inventario para obtener el valor total general. Finalmente muestra el inventario completo con la nueva información agregada

```python
import pandas as pd

productos = [
    {"codigo": "1", "Nombre": "Laptop", "Precio": 24000, "Cantidad": 10},
    {"codigo": "2", "Nombre": "Mochila para escuela", "Precio": 2000, "Cantidad": 4},
    {"codigo": "3", "Nombre": "Iphone 13", "Precio": 22000, "Cantidad": 2},
    {"codigo": "4", "Nombre": "Teclado Mecanico", "Precio": 1200, "Cantidad": 11},
    {"codigo": "5", "Nombre": "Mouse Logitech", "Precio": 1300, "Cantidad": 15}
]


inventario = pd.DataFrame(productos)


print("=== PRODUCTOS ===")
print(inventario)


codigo_Busqueda = "2"

producto_encontrado = inventario[inventario["codigo"] == codigo_Busqueda]

print("=== BUSQUEDA DE PRODUCTO ===")

if not producto_encontrado.empty:
    print(producto_encontrado)
else:
    print("Producto no encontrado")


inventario["Valor Total"] = inventario["Precio"] * inventario["Cantidad"]


valor_inventario = inventario["Valor Total"].sum()

print("=== VALOR TOTAL DEL INVENTARIO ===")
print(f"L {valor_inventario}")


print("=== INVENTARIO COMPLETO ===")
print(inventario)
```

# Ejercicio 14:   Registro de ventas

### Enunciado: cree una lista de ventas, cada venta debe tener cliente, producto, cantidad y precio unitario. Calcule el total por venta, calcule el total general vendido e identifique el cliente con la compra más alta.

### Lógica: el código crea un registro de ventas con clientes, productos, cantidades y precios, y lo organiza en una tabla usando pandas. Luego calcula el total de cada venta multiplicando la cantidad por el precio unitario y lo agrega como una nueva columna. Después muestra todas las ventas y suma el total general vendido. Finalmente identifica la venta más alta y muestra el cliente que realizó la compra de mayor valor junto con el monto

```python
import pandas as pd 

ventas = [
    {"Cliente": "Juan", "Producto": "Laptop", "Cantidad": 1, "Precio Unitario": 24000},
    {"Cliente": "Maria", "Producto": "Mouse", "Cantidad": 3, "Precio Unitario": 1200},
    {"Cliente": "Carlos", "Producto": "Teclado", "Cantidad": 2, "Precio Unitario": 1800},
    {"Cliente": "Ana", "Producto": "Monitor", "Cantidad": 1, "Precio Unitario": 5500},
    {"Cliente": "Pedro", "Producto": "Laptop", "Cantidad": 2, "Precio Unitario": 24000}
]


df_ventas = pd.DataFrame(ventas)


df_ventas["Total Venta"] = df_ventas["Cantidad"] * df_ventas["Precio Unitario"]


print("=== REGISTRO DE VENTAS ===")
print(df_ventas)


total_general = df_ventas["Total Venta"].sum()

print("=== TOTAL GENERAL VENDIDO ===")
print(f"L {total_general}")


cliente_mayor_compra = df_ventas.loc[df_ventas["Total Venta"].idxmax()]

print("=== CLIENTE CON LA COMPRA MÁS ALTA ===")
print("Cliente:", cliente_mayor_compra["Cliente"])
print("Total Compra:", cliente_mayor_compra["Total Venta"])
```

# Ejercicio 15:   Limpieza básica de texto

### Enunciado: defina o solicite una frase, elimine espacios al inicio y al final. Convierta el texto a minúsculas y reemplace espacios dobles por un solo espacio. Cuente cuántas palabras contiene y explique los métodos strip(), lower(), replace() y split().

### Lógica: el código toma una frase y la va limpiando paso a paso antes de analizarla. Primero elimina los espacios innecesarios al inicio y al final con strip(). Luego convierte todo el texto a minúsculas con lower() para estandarizarlo. Después intenta corregir espacios dobles usando replace(), dejando la frase más ordenada. Posteriormente usa split() para separar la frase en palabras y formar una lista, y con len() cuenta cuántas palabras hay en total. Finalmente muestra la frase limpia y la cantidad de palabras.

```python
frase = "  LA programación es divertida y gratificante  "


frase = frase.strip()


frase = frase.lower()


frase = frase.replace("  ", " ")


palabras = frase.split()
cantidad_palabras = len(palabras)


print("=== TEXTO LIMPIO ===")
print(frase)

print("\nCantidad de palabras:", cantidad_palabras)
```
